```python
import numpy as np
import matplotlib.pyplot as plt
from DynSystem import Dynamical_system
```

```python
def harmonic(x: np.array, t):
  return np.array([x[1], -np.sin(x[0])])

def damped_harmonic(x: np.array, t):
  return np.array([x[1], -np.sin(x[0])-0.01*x[1]])

def husband_wife(x: np.array, t):
  return np.array([x[1], -x[0]])

def rabbit_vs_sheep(x: np.array, t):
  return np.array([x[0]*(3-x[0]-2*x[1]), x[1]*(2-x[0]-x[1])])

def double_well(x: np.array, t):
  return np.array([x[1], x[0] - x[0]**3])
```

```python
# initail conditions
x0 = np.linspace(0, np.pi-0.1, 7, dtype=float)
v0 = np.zeros(7, dtype=float)
z0 = np.stack([x0, v0], axis=1).T

# dynamical system
harm_osc = Dynamical_system(harmonic, z0, 20, 500)
harm_osc.compute_sol()
harm_osc.plot_phase_portrait()
```

```python
# initial conditions
x0 = np.concatenate([np.linspace(0, np.pi-0.1, 7, dtype=float), np.ones(3, dtype=float)*np.pi])
v0 = np.concatenate([np.zeros(7, dtype=float), np.linspace(0.05, 0.5, 3, dtype=float)])
z0 = np.stack([x0, v0], axis=1).T

# dynamical system
damped_harm_osc = Dynamical_system(damped_harmonic, z0, 10, 500)
damped_harm_osc.compute_sol()
damped_harm_osc.plot_phase_portrait(Nx = 35, Ny = 12)
```

```python
# initail conditions
x0 = np.linspace(0, np.pi, 5, dtype=float)
y0 = x0
z0 = np.stack([x0, y0], axis=1).T

# dynamical system
wife_husb = Dynamical_system(husband_wife, init_cond=z0, tmax=20, N=500)
wife_husb.compute_sol()
wife_husb.plot_phase_portrait()
```

```python
# initail conditions
x0 = np.zeros(5)
y0 = np.linspace(0, 2, 5)
z0 = np.stack([x0, y0], axis=1).T

# dynamical system
doub_well = Dynamical_system(double_well, init_cond=z0, tmax=10, N=500)
doub_well.compute_sol()
#doub_well.plot_field(Nx=25, Ny=20, figsize=(12, 7))
doub_well.plot_phase_portrait()
```
