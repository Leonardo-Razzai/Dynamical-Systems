```python
from DynSystem import Dynamical_system
import numpy as np
```

For these examples see strogaz pag 181 

```python
# define dyn system
def van_der_pol_diff(z: np.array, t):
  return np.array([
    z[1],
    -z[0] + z[1]*(1 - z[0]**2)
  ])
  
# initial cond
x0 = np.zeros(6)
y0 = np.linspace(-2, 2, 6)
z0 = np.stack([x0, y0])
van_der_pol = Dynamical_system(van_der_pol_diff, z0, tmax = 10, N = 500)
van_der_pol.compute_sol()
van_der_pol.plot_phase_portrait(figsize=(12, 7))
van_der_pol.plot_x_t(init_cond_index=2)
```

```python
# define dyn system
def dipole_fixed_point_diff(z: np.array, t):
  return np.array([
    2*z[0]*z[1],
    z[1]**2 - z[0]**2
  ])
  
# initial cond
x0 = np.array([-0.01, 0.01])
y0 = np.array([0.5, 0.5])
z0 = np.stack([x0, y0])
dipole_fixed_point = Dynamical_system(dipole_fixed_point_diff, z0, tmax = 30, N = 10000)
dipole_fixed_point.compute_sol()
dipole_fixed_point.plot_phase_portrait(figsize=(12, 7))
```
